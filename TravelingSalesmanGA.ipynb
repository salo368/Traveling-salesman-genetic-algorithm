```python
import random
import string
import copy
import matplotlib.pyplot as plt
import numpy as np
import cv2
import io
import os
from IPython.display import clear_output

# Configurar el tamao de las imgenes y el nmero de imgenes
n_imagenes = 1000
alto, ancho = 300, 400

# Generar un vector de imgenes
imagenes = []


def apareamiento(v1,v2):
  first=v1[0]
  del v1[0]
  del v2[0]
  def mezclar(v1, v2,k):
    
    for i in range(k):
        temp = v1[i]
        index_v2i = v1.index(v2[i])
        v1[i] = v2[i]
        v1[index_v2i] = temp
    return v1
  k = random.randint(1, len(v1)-1)

  #mutacion
  h1=mezclar(v1,v2,k)
  h2=mezclar(v2,v1,k)
  if random.random() <= 0.5: #Probabilidad de mutacion (un intercambio)
    i, j = random.randint(0, len(h1)-1), random.randint(0, len(h1)-1)
    h1[i], h1[j] = h1[j], h1[i]
  if random.random() <= 0.5: #Probabilidad de mutacion (un intercambio)
    i, j = random.randint(0, len(h1)-1), random.randint(0, len(h1)-1)
    h2[i], h2[j] = h2[j], h2[i]

  h1.insert(0, first)
  h2.insert(0, first)



  return h1,h2

def ordenarADN(tupla1,ADNBase):
  #Pitagoras
  def dist(x0,y0,x1,y1):
    result=((x0-x1)**2+(y0-y1)**2)**0.5
    return result


 #ordenamiento burbuja, aqupi ordeno por el vector de distancias que esta
  #sicronizado con el ADN asi que se ordenan a la vez
  def ordenamiento_burbuja(lista,listaO):
      n = len(lista)
      auxiliar=[]
      auxiliar2=[]
      for i in range(n):
          for j in range(0, n-i-1):
              if lista[j] > lista[j+1]:
                  lista[j], lista[j+1] = lista[j+1], lista[j]
                  auxiliar=copy.copy(listaO[j])
                  auxiliar2=copy.copy(listaO[j+1])
                  for k in range(len(auxiliar)):
                    listaO[j][k]=auxiliar2[k]
                    listaO[j+1][k]=auxiliar[k]

  #recorre el ADN en busca de la suma de todas las distacias
  def distotal(ADNT):
    sumTemp= 0
    for j in range(len(ADNT)-1):  
      sumTemp= sumTemp+dist(tupla1[ADNT[j]][0],tupla1[ADNT[j]][1],tupla1[ADNT[j+1
      ]][0],tupla1[ADNT[j+1]][1])
    return sumTemp
  ADN2=[]
  ADNDist=[]

  #cierro el ADN
  for i in range(len(ADNBase)):
    v=[]
    for j in range(len(ADNBase[i])):
      v.append(ADNBase[i][j])
    v.append(ADNBase[i][0])
    ADN2.append(v)

  #Halla todas las distacias totales
  for i in range(len(ADN2)):
    ADNT=ADN2[i]
    ADNDist.append(distotal(ADNT))

  #ordeno el ADN
  ordenamiento_burbuja(ADNDist,ADN2)
  #print(ADNDist)
  for i in range(len(ADN2)):
    ADN2[i]=ADN2[i][:-1]
  #print(ADN2)
  return(ADN2,ADNDist)

#////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////////



def graficar (ciudades, enlace,gen,cada,d):
  enlace.append(enlace[0])
  x=[]
  y=[]
  name=[]
  id=[]
  colores= []
  for i in range (len(ciudades)):
    colores.append('k')
  colores[enlace[0]]='red'
    
  for i in range (len(ciudades)):
    x.append(ciudades[i][0])
    y.append(ciudades[i][1])
    name.append(ciudades[i][2])
    id.append(ciudades[i][3])

  # Graficar los puntos
  plt.scatter(x, y, c=colores)

  # Agregar nombres a los puntos
  for i in range(len(x)):
      plt.text(x[i]+0.1, y[i]+0.1, f' {name[i]}')

  # Enlazar algunos puntos con una lnea
  for i in range (len(x)):
    s=enlace[i]
    s1=enlace[i+1]
    if i== len(x)-1:
      plt.plot([x[s], x[s1]], [y[s], y[s1]], color='red')
    else:
      plt.plot([x[s], x[s1]], [y[s], y[s1]], color='k')
  plt.title('Generacion '+str(gen)+'- dist:'+str(d)[:7], fontsize=14)
  buf = io.BytesIO()
  plt.savefig(buf, format='png', dpi=100)
  buf.seek(0)
  img_array = np.frombuffer(buf.getvalue(), dtype=np.uint8)
  img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
  # Redimensionar la imagen
  img = cv2.resize(img, (ancho, alto))
  # Agregar la imagen al vector
  imagenes.append(img)
  # Mostrar la grfica
  
  plt.close()
  if(gen % cada==0):
    print("----------------------------------")
    plt.show()

#graficar(ciudades,vectores[0])

def generarCiudades(n):
  lugares = []

  # Creamos un bucle for para crear cada ciudad y agregarla a la lista de nodos
  for i in range(0,n):
      # Generamos valores aleatorios para las coordenadas x e y de la ciudad
      x = random.uniform(0, 100)
      y = random.uniform(0, 100)
      # Generamos un nombre aleatorio para la ciudad
      name = string.ascii_uppercase[i]
      # Creamos un objeto Nodo con los valores aleatorios y el nombre generado
      ciudad = Nodo(x, y, name, i)
      # Agregamos el objeto Nodo a la lista de nodos
      lugares.append(ciudad)
      

  ciudadesatr = []

  # Recorremos la lista de ciudades y guardamos sus atributos en un diccionario
  for ciudad in lugares:
      ciudad_con_atributos = {
          "x": ciudad.x,
          "y": ciudad.y,
          "name": ciudad.name,
          "id": ciudad.id
      }
      # Agregamos el diccionario a la lista de ciudades con atributos
      ciudadesatr.append(ciudad_con_atributos)

  
  ciudades = [[ciudad['x'], ciudad['y'], ciudad['name'], ciudad['id']] 
      for ciudad in ciudadesatr]
  return lugares,ciudades

class Nodo:
    def __init__(self, x: float, y: float, name: str, id: int):
        self.x = x
        self.y = y
        self.name = name
        self.id = id

def distADN(ADN_s,ciudades):
  distTotal=0
  def dist(x0,y0,x1,y1):
    result=((x0-x1)**2+(y0-y1)**2)**0.5
    return result
  for i in range(len(ADN_s)-1):
    distTotal=distTotal+dist(ciudades[ADN_s[i]][0],ciudades[ADN_s[i]][1],ciudades[ADN_s[i+1]][0],ciudades[ADN_s[i+1]][1])
  distTotal=distTotal+dist(ciudades[1][0],ciudades[1][1],ciudades[ADN_s[len(ADN_s)-1]][0],ciudades[ADN_s[len(ADN_s)-1]][1])
  return distTotal

#Empieza la ejecucion -----------------------------------------------------------------------------

n = int(input("Ingrese el numero de ciudades: "))

R=generarCiudades(n)
lugares=R[0]
ciudades=R[1]

for city in ciudades: 
  print("Ciudad",city[2],"- Coordenadas (",str(city[0])[:5],",",str(city[1])[:5],")- ID:",str(city[3]))

cinicial = input("Ingrese el id de la ciudad inicial: ")
ciudadd = lugares[int(cinicial)]
lugares[0], lugares[int(cinicial)] = lugares[int(cinicial)], lugares[0]


nvectores = 10
vectores = []
vector_set = set()  # conjunto para almacenar los vectores generados

while len(vectores) < nvectores:
    vector = lugares.copy()  # hacer una copia de la lista original de ciudades
    for j in range(1, len(lugares)):
        k = random.randint(1, len(lugares) - 1)  # elegir un ndice aleatorio para intercambiar
        vector[j], vector[k] = vector[k], vector[j]  # intercambiar las ciudades
    
    if tuple(vector) not in vector_set:  # verificar si el vector ya existe en el conjunto
        vector_set.add(tuple(vector))  # agregar el vector al conjunto
        vectores.append([ciudad.id for ciudad in vector])  # agregar el vector a la lista de vectores como una lista de ids

print("----------------------------------")


generados = int(input("Ingrese el numero de generaciones: "))

imprimir=input("Imprimir generaciones en la terminal: (y/n)")
if imprimir=="y":
  cada=1
else: 
  cada=999999

for gen in range(generados+1):
  if imprimir!="y":
    clear_output(wait=True)                                                    #<--------------------------------------------
  print("Cargando generaciones:",str((gen)/float(generados)*100)[:5],"% - Generacion:",str(gen))
  if(gen % cada==0):
    print("Generacion: ",gen)
    print(vectores)
  ordenados=ordenarADN(ciudades,vectores)
  if(gen % cada==0):
    print(ordenados[0])
    print(ordenados[1])
  
  childs1=apareamiento(ordenados[0][0].copy(),ordenados[0][1].copy())
  childs2=apareamiento(ordenados[0][0].copy(),ordenados[0][1].copy())
  childs3=apareamiento(ordenados[0][2].copy(),ordenados[0][3].copy())
  childs4=apareamiento(ordenados[0][0].copy(),ordenados[0][2].copy())
  childs5=apareamiento(ordenados[0][1].copy(),ordenados[0][3].copy())
  nuevaGen=[ordenados[0][0].copy(),childs1[0],childs1[1],childs2[0],childs2[1],childs3[0],childs3[1],childs4[0],childs4[1],childs5[0]]
  
  
  d=distADN(ordenados[0][0],ciudades)
  graficar(ciudades,ordenados[0][0],gen,cada,d)
  vectores=nuevaGen.copy()

#-----------------------------------------------------------------------------------

print("Preparando Video")
# Convertir el vector de imgenes en un video
video = cv2.VideoWriter('GraficaVideo.avi', cv2.VideoWriter_fourcc(*'DIVX'), 10, (ancho, alto))
for imagen in imagenes:
    video.write(imagen)
video.release()
print("Video Listo")

#-------------------------------------------------------------------------------------------------

import os
from IPython.display import clear_output

# Especifica la ruta del archivo de video a borrar
ruta_del_video = '/content/GraficaVideo.mp4'

# Usa la funcin remove() de la librera os para borrar el archivo de video
try:
    os.remove(ruta_del_video)
    print(f"El archivo de video {ruta_del_video} ha sido eliminado exitosamente.")
except OSError as e:
    print("Ha ocurrido un error al eliminar el archivo de video:", e)

#----------------------------------------------------------------------------

# Especifica la ruta del archivo AVI
input_file = '/content/GraficaVideo.avi'

# Especifica la ruta y el nombre del archivo MP4 de salida
output_file = '/content/GraficaVideo.mp4'

# Ejecuta el comando ffmpeg para convertir el archivo AVI en MP4
!ffmpeg -i "{input_file}" "{output_file}"
if imprimir!="y":
 clear_output(wait=True)                                       #<--------------------------------------------
# Verifica que el archivo MP4 de salida se haya creado
if os.path.isfile(output_file):
    print(f"Archivo MP4 creado en {output_file}")
else:
    print("Ha ocurrido un error al crear el archivo MP4")

#--------------------------------------------------------------------------------------------------------------
from IPython.display import HTML
from base64 import b64encode

# Especifica la ruta del archivo MP4
ruta_del_video = '/content/GraficaVideo.mp4'

# Codifica el archivo MP4 como un objeto de datos base64
video_codificado = b64encode(open(ruta_del_video, 'rb').read()).decode('ascii')

# Muestra el archivo MP4 en un reproductor de video
HTML(f"""
<video width="640" height="360" controls>
  <source src="data:video/mp4;base64,{video_codificado}" type="video/mp4">
</video>
""")
```
